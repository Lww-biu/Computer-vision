```python
from PIL import Image
from pylab import *

im = array(Image.open('../img/camera.jpg'))
```

```python
#绘制图像
imshow(im)

#标记点
x = [100,100,400,400]
y = [200,500,200,500]
plot(x,y,'r*')

#绘制连线
plot(x[:2],y[:2])

#添加标题并显示
title('Plotting:"empire.jpg"')

#不显示坐标轴
axis('off')

show()
```

```python
im = array(Image.open('../img/camera.jpg').convert('L'))

# 新建一个图像
figure()
# 不使用颜色信息
gray()
# 在原点的左上角显示轮廓图像
contour(im, origin='image')
axis('equal')
axis('off')
```

```python
#直方图
figure()
hist(im.flatten(),128)
show()
```
